```python
#headers = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36 Edg/96.0.1054.62'}  
for i in range (1,50):
    c21Web =  requests.get('https://www.century21albania.com/en/properties?display=grid&city=Tirana&page=' + str(i))
    soup = BeautifulSoup(c21Web.content, "html.parser")
    results = soup.find_all('div', {'class' : 'card card-list'})
    results
```

```python
    results
```

```python
    location = [result.find('h6', {'class':'card-subtitle mt-1 mb-0 text-muted'}).get_text() for result in results]
```

```python
    description = [result.find('h5', {'class':'card-title'}).get_text() for result in results]
```

```python
    price = [result.find('h2', {'class':'text-primary mb-2'}).get_text() for result in results]
```

```python
    surface = [result.find('div', {'class':'col-xs-3 col-sm-3 col-md-3 col-lg-3 FutureInfo col-3'}).get_text() for result in results]
```

```python
    saleORrent = [result.find('div', {'class':'card-img'}).get_text() for result in results]
```

```python
    RE_dataset = pd.DataFrame(columns=['Location', 'Description', 'Price', 'Surface','Listing'])
    for j in range (len(location)):
        RE_dataset = RE_dataset.append({'Location':location[j], 'Description':description[j], 'Price':price[j], 'Surface':surface[j], 'Listing':saleORrent[j]}, ignore_index=True)
```

```python
RE_dataset
```

```python
RE_dataset ['Price'] = RE_dataset['Price'].apply(lambda line: line.replace('\n', ''))
RE_dataset ['Surface'] = RE_dataset['Surface'].apply(lambda line: line.replace('\n', ''))
RE_dataset ['Location'] = RE_dataset['Location'].apply(lambda line: line.replace('\n', ''))
RE_dataset ['Listing'] = RE_dataset['Listing'].apply(lambda line: line.replace('\n', ''))
```

```python
RE_dataset ['Company'] = 'Century21'
```
